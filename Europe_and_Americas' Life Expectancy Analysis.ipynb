```python
data = pd.read_csv('countries.csv')
data.head()
```

```python
# storing the 1997 dataset to a variable and selecting  America's and Europe

data_1997 = data[data.year == 1997]
americas_1997 = data_1997[data_1997.continent == 'Americas']
europe_1997 = data_1997[data_1997.continent == 'Europe']
```

```python
#to look at the set of unique values under asia_2007 dataset
set(americas_1997.country)
```

```python
#to look at the set of unique values under europe_2007 dataset
set(europe_1997.country)
```

```python
# using len to return the number of items in an object
print(len(set(americas_1997.country)))
print(len(set(europe_1997.country)))
```

```python
#finding the mean and median GDP per Capita in AMERICA and EUROPE

print('Mean Life Expectancy in Americas:')
print(americas_1997.lifeExpectancy.mean())
print('Mean Life Expectancy in Europe: ')
print(europe_1997.lifeExpectancy.mean())
print('Median Life Expectancy in Americas:')
print(americas_1997.lifeExpectancy.median())
print('Median Life Expectancy in Europe: ')
print(europe_1997.lifeExpectancy.median())
```

### CREATING HISTOGRAM

```python

# This means we want a canvas that contains
#(2 rows, 1 column , and index of the function we want to select the top to show the AMERICAS HISTOGRAM)


bins = 20
ax = plt.subplot(211)
plt.title('Distribution of Life Expectancy')
plt.hist(americas_1997.lifeExpectancy, bins, edgecolor='black', range=(55, 85)) #adjusted the range to match and properly compare the two


# This means we want a canvas that contains
#(2 rows, 1 column , and index of the function we want to select the top to show the EUROPE HISTOGRAM)
plt.subplot(212, sharex=ax, sharey=ax)
plt.hist(europe_1997.lifeExpectancy, bins, edgecolor='black', range=(55, 85))
plt.ylabel('Europe')
plt.show()
```

- By comparing the two histograms shows that there are many more countries on the higher end in terms of life expectancy in Europe than in Americas.

- There are also few outliers that have a very low life expectancy in Americas.

- To find these two outliers, we can filter them using the code below

```python
americas_1997[americas_1997.lifeExpectancy < 65]
```
