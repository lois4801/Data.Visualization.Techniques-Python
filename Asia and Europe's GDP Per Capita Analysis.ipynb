### We want to select the data from Asia and Europe in 2007

```python
# Select the data from year 2007 and store in its own variable

data_2007 = data[data.year == 2007]
```

```python
# storing the Asia and Europe Data in their respective variable

asia_2007 = data_2007[data_2007.continent == 'Asia']
europe_2007 = data_2007[data_2007.continent == 'Europe']
```

```python
#taking the first 5 on the dataset
asia_2007.head()
```

```python
# taking the first 5 on the dataset
europe_2007.head()
```

```python
#to look at the set of unique values under asia_2007 dataset
set(asia_2007.country)
```

```python
#to look at the set of unique values under europe_2007 dataset
set(europe_2007.country)
```

```python
# using len to return the number of items in an object

print(len(set(asia_2007.country)))
print(len(set(europe_2007.country)))
```

```python
#finding the mean and median GDP per Capita in ASIA and Europe

print('Mean GDP Per Capita in Asia:')
print(asia_2007.gdpPerCapita.mean())
print('Mean GDP Per Capita in Europe:')
print(europe_2007.gdpPerCapita.mean())
print('Median GDP Per Capita in Asia:')
print(asia_2007.gdpPerCapita.median())
print('Median GDP Per Capita in Europe:')
print(europe_2007.gdpPerCapita.median())
```

### Creating Histogram

```python
#Creating histogram
import matplotlib.pyplot as plt

plt.hist(asia_2007.gdpPerCapita)
plt.show
```

```python
#Creating histogram
import matplotlib.pyplot as plt


# This means we want a canvas that contains (2 rows, 1 column , and index of the function we want to select the top to show the ASIA HISTOGRAM)
plt.subplot(2, 1, 1) # subplot(211)
plt.title('Distribution of GDP Per Capita')
plt.hist(asia_2007.gdpPerCapita, 20, range=(0, 50000), edgecolor='black') #added range 50000 to match and compare both charts
plt.ylabel('Asia')

# This means we want a canvas that contains (2 rows, 1 column , and index of the function we want to select the bottom to show EUROPE HISTOGRAM)
plt.subplot(2, 1, 2)
plt.hist(europe_2007.gdpPerCapita, 20, range=(0, 50000), edgecolor='black')
plt.ylabel('Europe')
plt.show()
```
