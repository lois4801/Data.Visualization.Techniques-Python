### DATA AGGREGATION EXAMPLE
- A scatter plot of approval ratings
- mean and median of approval ratings

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('obama.csv', parse_dates=['year_month'])
data.head()
```

```python
#the "o" means that I want to plot the points only as small circles(no lines between them) with a size of 2
# the x axis= data.year_month
#the y axis= data.approve_percent
# the alpa=0.3 is to make the data points transparent to get a better idea on where the points are clustered

plt.plot(data.year_month, data.approve_percent, 'o', markersize=2,
         alpha=0.3)
plt.show()
```

```python
#aggregate the data and summarize it

# group by year month and get the mean and median of it
data_mean = data.groupby('year_month').mean()
data_median = data.groupby('year_month').median()

#data.mean index and data.median index give us the the year_month column and
plt.plot(data_mean.index, data_mean.approve_percent, 'red')
plt.plot(data_median.index, data_median.approve_percent, 'green')

plt.legend(['Mean', 'Median'])
plt.plot(data.year_month, data.approve_percent, 'o', markersize=2,
         alpha=0.3)
plt.show()
```

Checking for the 75th and 25th percentiles of approval ratings for each month


```python

data_25 = data.groupby('year_month').quantile(0.25)
data_75 = data.groupby('year_month').quantile(0.75)

plt.plot(data_75.index, data_75.approve_percent, 'green')
plt.plot(data_median.index, data_median.approve_percent, 'red')
plt.plot(data_25.index, data_25.approve_percent, 'blue')

plt.plot(data.year_month, data.approve_percent, 'o', markersize=2,
         alpha=0.3)

plt.legend(['75th Percentile', 'Median', '25th Percentile'])
plt.show()
```

SAMPLING METHOD TECHNIQUES
- What to do when your data is too big
 - Aggregate data ( mean, percentiles, etc)
 - Sample a random subset of data
 - If the website has 100million users in the website it may not be necessary to keep track all their  behaviors on the website. It is safe to say that 10 percent of the users will be used only for data analysis purposes.
