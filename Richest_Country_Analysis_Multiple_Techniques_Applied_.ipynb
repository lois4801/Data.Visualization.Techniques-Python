### What is the richest country in the world?
- Applying the different techniques
  - Aggregating values with groupby()
  - Soring values
  - Making line charts/time series
  - Using subplots


```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('countries.csv')
data.head()
```

```python
# checking the countries mean for each attributes
data.groupby(['country']).mean()
```

```python
#finding the mean value for the gdp_per_capita

mean_gdp_per_capita = data.groupby(['country']).mean().gdpPerCapita
mean_gdp_per_capita
```

```python
#sorting the mean values by descending and showing the top 5 only

top5 = mean_gdp_per_capita.sort_values(ascending=False).head()
top5
```

# Is Kuwait really the richest country in the world (on a per-person basis)?

```python
# Showing the top 5 data from Kuwait

kuwait = data[data.country == 'Kuwait']
kuwait.head()
```

```python
plt.plot(kuwait.year, kuwait.gdpPerCapita)
plt.title('GDP Per Capita in Kuwait')
plt.show()
```

- There is a massive drop of GDP Per Capita in Kuwait between 1970 to 1990.  To check what happened I will going to do subplot for population, gdp per capita and gdp in billions.

```python
# 311 means - 3 rows, 1 column, and 1st top chart

plt.subplot(311)
plt.title('GDP Per Capita')
plt.plot(kuwait.year, kuwait.gdpPerCapita)

plt.subplot(312)
plt.title('GDP in Billions')
plt.plot(kuwait.year, kuwait.population*kuwait.gdpPerCapita / 10**9)

plt.subplot(313)
plt.title('Population in Millions')
plt.plot(kuwait.year, kuwait.population / 10**6)

#to fix the title of the layout
plt.tight_layout()
plt.show()
```

```python
#showing the population growth

plt.plot(kuwait.year, kuwait.population / kuwait.population.iloc[0] * 100)

kuwait_gdp = kuwait.population * kuwait.gdpPerCapita
plt.plot(kuwait.year, kuwait_gdp / kuwait_gdp.iloc[0] * 100)
plt.plot(kuwait.year, kuwait.gdpPerCapita / kuwait.gdpPerCapita.iloc[0] * 100)
plt.title('GDP and Population Growth in Kuwait (first year = 100)')
plt.legend(['Population', 'GDP', 'GDP Per Capita'])
plt.show()
```
