```python
data = pd.read_csv('countries.csv')
data.head()
```

```python
#selecting the all the data under country which is United States

us = data[data.country == 'United States']
us.head()
```

### Line Graph for US


```python
#Doing now the line graph for US

plt.plot(us.year, us.gdpPerCapita)  # this would be the (X axis, Y axis)

plt.title('GDP Per Capita Growth in the U.S.')
plt.xlabel('year')

plt.ylabel('GDP per capita')
plt.show()
```

The graph shows that there was about 15000 dollars per year in 1950's , and it grows to more than $40,000 dollars in 2000's  

```python
#selecting the all the data under country which is China

china = data[data.country == 'China']
```

```python
### Line Graph for China

plt.plot(us.year, us.gdpPerCapita) # for the top
plt.plot(china.year, china.gdpPerCapita) #for the bottom

plt.legend(['United States', 'China']) # using legend to properly identify which line for each country

plt.xlabel('year')
plt.ylabel('GDP per capita')
plt.show()
```

```python
# Instead of comparing the absolute values, we will compare the growth instead.
# Let's set the GDP per capita in the first available year to 100 and compare how it grew in US and Chine for each 10 years


# Iloc is integer location and this is used to select an item with an integer index

#This is US GDP per Capita every year.

us.gdpPerCapita
```

```python
# the row column 1608 is equivalent to 1952 in the year column
us.gdpPerCapita.iloc[0]
```

```python
#using the formula below to see the growth for all year

us_growth = us.gdpPerCapita / us.gdpPerCapita.iloc[0] * 100
us_growth
```

```python
china_growth = china.gdpPerCapita / china.gdpPerCapita.iloc[0] * 100
china_growth
```

```python


plt.title('GDP Per Capita Growth (first year = 100)')

plt.plot(us.year, us_growth)  # plotting it by growth instead of like this  plt.plot(us.year, us.gdpPerCapita)
plt.plot(china.year, china_growth) # plotting it by growth instead of like this  plt.plot(china.year, china.gdpPerCapita)

plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('GDP per capita growth')
plt.show()
```

- This graph now shows that the GDP per capita in China has grown much faster than that of the of the United states.
